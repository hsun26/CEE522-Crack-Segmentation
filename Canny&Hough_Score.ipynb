```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
image_path = './test_new'  
images = []
for filename in os.listdir(image_path):
    img = mpimg.imread(os.path.join(image_path, filename))
    images.append(img)
#plt.imshow(images[8])
```

```python
shape = np.shape(images[8])
```

```python
segmented_images = []
for img in images:
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  
    kernel_size = 5
    blur_gray = cv2.GaussianBlur(img,(kernel_size, kernel_size), 0)
    low_threshold = 50
    high_threshold = 150
    edges = cv2.Canny(blur_gray, low_threshold, high_threshold)
    rho = 2 
    theta = np.pi/180 
    threshold = 15     
    min_line_length = 30 
    max_line_gap = 20    
    line_image = np.copy(img)*0 
    lines = cv2.HoughLinesP(edges, rho, theta, threshold, np.array([]),
                            min_line_length, max_line_gap)      
    if lines is not None:
        for line in lines:
            for x1,y1,x2,y2 in line:
                cv2.line(line_image,(x1,y1),(x2,y2),(255,0,0),10)
    
    color_edges = np.dstack((edges, edges, edges)) 
 
    lines_edges = cv2.addWeighted(color_edges, 0.8, line_image, 1, 0) 
 

    '''lines = cv2.HoughLines(edges, 1, np.pi/180, 200) 
    line_img = np.zeros_like(img)  
    if lines is not None:
        for line in lines:
            rho, theta = line[0]
            a = np.cos(theta)
            b = np.sin(theta)
            x0 = a * rho
            y0 = b * rho
            x1 = int(x0 + 1000 * (-b))
            y1 = int(y0 + 1000 * (a))
            x2 = int(x0 - 1000 * (-b))
            y2 = int(y0 - 1000 * (a))
            cv2.line(line_img, (x1, y1), (x2, y2), (0, 0, 255), 2) 
    ''' 
    segmented_img = lines_edges
    segmented_images.append(segmented_img)
#plt.imshow(segmented_images[8])
```

```python
segmented_images = np.array(segmented_images)  
np.shape(segmented_images) 
```

```python
ground_truth_path = './masks_new'  
gt_images = []  
for filename in os.listdir(ground_truth_path):
    mask = cv2.imread(os.path.join(ground_truth_path, filename))
    gt_images.append(mask)
```

```python
gt_images = np.array(gt_images)
np.shape(gt_images)
```

```python
images_new = []
for image in segmented_images:
    image_new = np.array(image)
    _, image_new = cv2.threshold(image_new, 127, 255, cv2.THRESH_BINARY)
    image_new = image_new/255
    images_new.append(image_new)
```

```python
for image in images_new:
    print(np.unique(image))
```

```python
masks_new = []
for image in gt_images:
    mask_new = np.array(image)
    _, mask_new = cv2.threshold(mask_new, 127, 255, cv2.THRESH_BINARY)
    mask_new = mask_new/255
    masks_new.append(mask_new)
```

```python
for image in masks_new:
    print(np.unique(image))
```

```python
def dice_coef_metric(pred, label):
    intersection = 2.0 * (pred * label).sum()
    union = pred.sum() + label.sum()
    
    if pred.sum() == 0 and label.sum() == 0:
        return 1
    return intersection / union
```

```python
def IoU_metric(pred, label):
    intersection = np.logical_and(pred, label)
    union = np.logical_or(pred, label)
    iou_score = intersection.sum() / union.sum()
    if pred.sum() == 0 and label.sum() == 0:
        return 1
    return iou_score
    
```

```python
dice_scores = []  
iou_scores = []  
for i in range(len(images)):
    image_new = images_new[i]
    mask_new = masks_new[i]
    dice_score = dice_coef_metric(image_new, mask_new)
    dice_scores.append(dice_score)
    iou_score = IoU_metric(image_new, mask_new)
    iou_scores.append(iou_score)
```

```python
print("Mean Dice score:", np.mean(dice_scores))
print("Mean IoU score:", np.mean(iou_scores))
print("Mean Dice score:", dice_scores)
print("Mean IoU score:", iou_scores)
```
